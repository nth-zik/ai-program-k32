```
import pathlib
import os

data_test = "/content/drive/MyDrive/Landmark_Classification/test"
data_train = "/content/drive/MyDrive/Landmark_Classification/train"
folders_test = os.listdir(data_test)
folders_train = os.listdir(data_train)
print(folders_test)
print(folders_train)
```

```
import cv2
from imutils import paths
import numpy as np

image_train=[]
image_labels=[]
image_names=[]
size=180,180
for(i, imgpath) in enumerate(list(paths.list_images(data_train))):
  print(i, imgpath)
  img=cv2.imread(imgpath)
  label=imgpath.split('/')[-2]
  img = cv2.resize(img, size)
  image_train.append(img)
  image_labels.append(label)

X=np.array(image_train)
Y=np.array(image_labels)
```

```
import cv2
from imutils import paths
import numpy as np

image_test=[]
image_labels_test=[]
image_names_test=[]
size=180,180
for(i, imgpath) in enumerate(list(paths.list_images(data_test))):
    print(i, imgpath)
    img=cv2.imread(imgpath)
    label=imgpath.split('/')[-2]
    img = cv2.resize(img, size)
    image_test.append(img)
    image_labels_test.append(label)#Train — Test Split

X_val=np.array(image_test)
Y_val=np.array(image_labels_test)
```

```
from sklearn.preprocessing import LabelEncoder
import keras
from keras.utils.np_utils import to_categorical
from sklearn.model_selection import train_test_split

print(len(folders_train))
le=LabelEncoder()
Y_transformed=le.fit_transform(Y)
Y_transformed=to_categorical(Y_transformed, num_classes=len(folders_train))

leval=LabelEncoder()
Y_valtransformed=leval.fit_transform(Y_val)
Y_valtransformed=to_categorical(Y_valtransformed, num_classes=len(folders_train))

# Chuyển thành dạng (số lượng ảnh, số kênh * chiều cao * chiều rộng)
X1 = X.reshape(X.shape[0], 3 * size[0] * size [1] )
X_val1 = X_val.reshape(X_val.shape[0], 3 * size[0] * size [1] )
# train
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X1, Y_transformed)
acc = model.score(X_val1, Y_valtransformed)
print("Raw pixel accuracy: {:.2f}%".format(acc * 100))
print(classification_report(Y_valtransformed, model.predict(X_val1), target_names=le.classes_))
```
